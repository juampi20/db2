## Ejercicio 1

- Descomprimir el archivo clima.tgz de la carpeta data
- Convertir cada archivo a Dataframe y mostrarlo en pantalla

```python
import tarfile
import pandas as pd
import os

# Constant
path = './data/clima/'

# Descomprimir el archivo clima.tgz de la carpeta data
tar = tarfile.open("data/clima.tgz", "r")
tar.extractall("data/clima/")
tar.close()

# Convertir cada archivo a Dataframe y mostrarlo en pantalla
for file in os.listdir(path):
    if file.endswith(".txt"):
        df = pd.read_csv(os.path.join(path, file), encoding='ISO-8859-1', sep="\t")
        print(f"\nFile: {file}")
        print(df)
```

## Ejercicio 2

- Obtener de la API de <a href="https://wikimedia.org/api/rest_v1/#/Pageviews%20data">Wikimedia </a> la cantidad de visitas en este año a la página Albert Einstein por día.
- Mostrar el Dataframe obtenido por pantalla

```python
import pandas as pd
import urllib.request
import json

URL = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia.org/all-access/all-agents/Albert_Einstein/daily/20210101/20210829'

# Get json from web
with urllib.request.urlopen(URL) as url:
    data = json.loads(url.read())

# Collect total views
total_views = 0
for item in data['items']:
    total_views += item['views']

# Create DataFrame for data
df = pd.DataFrame(data['items'])

# Show Table
print(df[['project', 'article', 'granularity', 'timestamp', 'views']])
print(f"\nTotal Views: {total_views}")
```

## Ejercicio 3

- Crear una base de datos MySQL
- Crear una tabla que contenga nombre, apellido y teléfono
- Obtener de la base de datos la lista de personas ordenada por apellido de forma descendente
- Mostrar el Dataframe resultante por pantalla

```python
import pandas as pd
import sqlite3

# Open file .db
con = sqlite3.connect('./data/course.db')
con.execute("PRAGMA busy_timeout = 30000")

# Create Object Cursor
c = con.cursor()
c.execute('create table if not exists person(firstname, lastname, phone)')

# Insert Values
c.execute('''insert into person(firstname, lastname, phone) values
('Oscar', 'Zeta', '2604569531'),
('Juampi', 'Garcia', '2625432584'),
('Carlos', 'Rodriguez', '2648521816'),
('Ezequiel', 'Garcia', '2625464355'),
('Lucas', 'Arbeloa', '2665489532')''')

# Get Table order by lastname
res = c.execute('select * from person order by "lastname"')

# Show data
df = pd.DataFrame(res.fetchall())
print(df)

# Close connection
con.close()
```

## Ejercicio 4

- Obtener de la API <a href="https://date.nager.at/swagger/index.html">Nager.Date</a> las fiestas nacionales argentinas para el 2021
- Mostrar el dataframe resultante por pantalla

```python
import pandas as pd
import json
import requests

URL = 'https://date.nager.at/api/v2/PublicHolidays/2021/AR'

with requests.get(URL) as url:
    data = json.loads(url.text)

df = pd.DataFrame(data)
print(df[['date', 'localName']])
```

## Ejercicio 5

- Obtener una tabla informativa de una página a elección con Web Scrapping
- Mostrar el Dataframe resultante por pantalla

```python
import pandas as pd
import requests
import json

URL = 'https://age-of-empires-2-api.herokuapp.com/api/v1/civilizations'

with requests.get(URL) as url:
    data = json.loads(url.text)

df = pd.DataFrame(data['civilizations'])
print(df[['name', 'expansion', 'army_type']])
```
