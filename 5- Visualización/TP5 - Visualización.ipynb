Convendria realizar otro grafico ya que contiene demasiados elementos y la sumatoria de estos elementos con respecto a sus porcentajes supera el 100%. Estos porcentajes de exitos son independiente de cada jugador de su propio 100%

Por este motivo no seria un grafico de torta sino de barras para mostrar el porcentaje de su 100%

Representando en el rango de y hasta 35 ya que ningun jugador pasa el 35%


```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(15,10))
x = ["Williamson","Rohit","Shakib","Warner","Babar","Su Plessis","Pooran","Root","Kusal Perera","Rahmat Shah"]
y = [30.23,29.05,28.25,25.02,24.51,21.06,20.01,19.07,18.16,14.8]
plt.bar(x,y)
plt.xlabel("guners")
plt.ylabel("Percentage")
plt.title("The world cup's big guns")
plt.show();
```

Este grafico no esta mal representado con barras pero se podria representar con una de serie de tiempo, la cual representaria los años que pasaron y como incremento su capital durante los años

Es medio confuso este grafico ya que utiliza colores similares y todo en un grafico de barras para demostrar dos cosas diferentes en este caso aplicamos el grafico de dispersion de puntos para marcar las muertes en diferentes ciudades con respecto a KWTS y AVE como se puede ver a continuacion.

```python
plt.figure(figsize=(15,10))
ax = ["Jasprit", "Rashid", "Liam", "Hasan", "Mustafizur", "Bhuvneshwar", "Pat", "Adil","Yuzvendra","Tendai"]
ay = [37,30,29,24,23,21,20,20,19,19]
bx = ["Jasprit", "Rashid", "Liam", "Hasan", "Mustafizur", "Bhuvneshwar", "Pat", "Adil","Yuzvendra","Tendai"]
by = [14.48,10.63,12.20,19.87,17.43,29.09,15.65,20.55,13.89,20.31]

plt.scatter(ax,ay,color='red',label='WKTS') 
plt.scatter(bx,by,color='blue',label='AVE') 
plt.xlabel("Cities")
plt.ylabel("Death")
plt.title("Deaths May 14, 2019")
plt.legend(loc='best') 
plt.show()
```
